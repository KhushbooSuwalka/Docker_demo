```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("DemoSparkApp").getOrCreate()
```

```python
data = [1,2,3,4,5]
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
for element in rdd.collect():
    print(element)
```

```python
spark.stop()
```
