```python
from kafka import KafkaProducer
import json

producer = KafkaProducer(
    bootstrap_servers = "kafka:9092",
    value_serializer = lambda x: json.dumps(x).encode("utf-8")
)
# value_serializer :- This converts Python objects (which are in json format) into bytes (binary format) because Kafka can only send bytes.
# json.dumps(x) :- converts dictionary to json string
# .encode("utf-8") :- Converts string into bytes. 

for i in range(10):
    producer.send(
        "demo-topic",
        {"id": i,"name":f"Student - {i}"}
    )

producer.flush()
# flush() ensures All pending messages are sent.Program waits until Kafka confirms receipt.Without it, the program might exit before all messages are delivered.
print("Done")
```
